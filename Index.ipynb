```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

# Index

```python
index_tickers = ['^GSPC', '^IXIC', '^GDAXI','^DJI']
```

```python
index_data = pd.DataFrame()

for t in index_tickers:
    index_data[t] = wb.DataReader(t, data_source='yahoo', start='2010-1-1')['Adj Close']
```

^GSPC:SPX500  ^IXIC:NASDAQ  ^GDAXI:DAX  ^DJI:DOW

```python
index_data.columns = ['SPX','NASDAQ','DAX','DOW']
```

```python
index_data.tail()
```

Normalize returns

```python
index_normalized = index_data/index_data.iloc[0]
index_normalized.plot(figsize = (15,6))
plt.show()
```

```python
index_returns = index_data.pct_change(1)
index_annual_returns = index_returns.mean() *250
round(index_annual_returns*100,2)
```
